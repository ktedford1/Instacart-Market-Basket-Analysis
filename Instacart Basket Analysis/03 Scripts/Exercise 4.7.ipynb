### 'If' statements and functions

```python
# define a function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# apply the function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# check it
df.head()
```

```python
# check the shape
df.shape
```

```python
# check the labels
df['price_range'].value_counts(dropna = False)
```

```python
# find the max price
df['prices'].max()
```

### 'If' statements using 'loc()' functions

```python
# create a new column, price_range_loc, using a loc function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# now apply it all on the complete df, df_ords_prods_merged

df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df_ords_prods_merged['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# check the value counts
df_ords_prods_merged['price_range_loc'].value_counts(dropna = False)
```

### 'If' statements using 'for' loops

```python
# what is the busiest day of the week (find the value counts for one column, 'orders_day_of_week')

df_ords_prods_merged['orders_day_of_week'].value_counts(dropna = False)
```

```python
# add the 'for' loop to label the days of the week

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# check it
result
```

```python
# create a new column in the df to hold the results
df_ords_prods_merged['busiest_day'] = result
```

```python
# check the value counts
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

#### Q2: Change “Busiest day” to “Busiest days” (plural). This label should correspond with the two busiest days of the week. Also, what are the two slowest days. 

The 2 busiest days are 0 and 1: Saturday and Sunday.
The 2 slowest days are 3 and 4: Tuesday and Wednesday.
Therefore, change the for loop from above to include these new days, Sunday [1] and Tuesday [3].

```python
# add the 'for' loop to label the days of the week - include the 2 extra days

result = []

for value in df_ords_prods_merged["orders_day_of_week"]:
  if (value == 0) or (value == 1):
    result.append("Busiest day")
  elif (value == 3) or (value == 4):
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# create a new column in the df to hold the results
df_ords_prods_merged['busiest_day'] = result
```

```python
# check the value counts
df_ords_prods_merged['busiest_day'].value_counts(dropna = False)
```

#### Q4: Identify the busiest hours of the day. Label periods of time “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# what is the busiest time of day
df_ords_prods_merged['orders_time'].value_counts(dropna = False)
```

```python
# find the average hour from the 'orders_time' column
average_orders_time = df_ords_prods_merged['orders_time'].mean()
print("Average of orders_time:", average_orders_time)
```

Note: A quick visual inspection of the value_counts from 'orders_time' shows that these fall into 3 groups - the 5-digits, the 6-digits, and the 7-digit sums of orders by hour. 
Therefore, I will label the 7-digit group (8-19) as the "Most orders", the 6-digit group (20-23, 0-1) as the "Average orders" and the 5-digit group (2-5) as the "Fewest orders". 

```python
# add the 'for' loop to label the hours of the day by the 3 groups

result2 = []

for value in df_ords_prods_merged["orders_time"]:
  if (value >= 8) and (value <= 19):
    result2.append("Most orders")
  elif ((value >= 20) and (value <= 23)) or (value == 0) or (value == 1) or (value == 6) or (value == 7):
    result2.append("Average orders")
  else:
    result2.append("Fewest orders")
```

```python
# create a new column in the df to hold the results
df_ords_prods_merged['busiest_period_of_day'] = result2
```

```python
# check the value counts
df_ords_prods_merged['busiest_period_of_day'].value_counts(dropna = False)
```
