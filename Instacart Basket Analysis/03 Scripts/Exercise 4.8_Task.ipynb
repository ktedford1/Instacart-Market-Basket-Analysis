#### Q4: Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# aggregate the data using transform
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# using a loc to flag customer loyalty - part 1
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
# using a loc to flag customer loyalty - part 2
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
# using a loc to flag customer loyalty - part 3
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# check the value counts of the new column "loyalty_flag"
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# check the top 5 rows
ords_prods_merge.head()
```

#### Q5: Use the loyalty flag you created and check the basic statistics of the product prices for each loyalty category (Loyal Customer, Regular Customer, and New Customer). 

```python
# use groupby and agg functions
ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Q5 answer:
The average product price decreases as the "loyalty" (total order numbers per customer) increases.

#### Q6: Create a spending flag for each user based on the average price across all their orders. If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”, equal or higher than 10, "High spender"

```python
# aggregate the data using transform to get the mean of product prices per customer_id, transformed in a new column
ords_prods_merge['spending_mean'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# using a loc to flag mean customer spending - part 1
ords_prods_merge.loc[ords_prods_merge['spending_mean'] >= 10, 'spending_flag'] = 'High Spender'
```

```python
# using a loc to flag mean customer spending - part 1
ords_prods_merge.loc[ords_prods_merge['spending_mean'] < 10, 'spending_flag'] = 'Low Spender'
```

```python
# check the value counts of the new column ('spending_flag')
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

#### Q7: Create a spending flag for each user based on the average price across all their orders.. find the median of days since prior order and flag it according to <=10 (non-frequent), 10<=20 (regular), >20 (frequent)

```python
# create a new column and calculate the median 'days_since_prior_order' for each user_id
ords_prods_merge['user_frequency'] = ords_prods_merge.groupby('user_id')['days_since_prior_order'].median()
```

```python
ords_prods_merge.head()
```

```python
# if median days_since_prior_order > 20 - label "non-frequent customer"
ords_prods_merge.loc[ords_prods_merge['user_frequency'] > 20, 'spend_status'] = 'Non-frequent customer'
```

```python
# median days_since_prior_order between 10 and 20 - label "Regular customer"
ords_prods_merge.loc[(ords_prods_merge['user_frequency'] > 10) & (ords_prods_merge['user_frequency'] <= 20), 'spend_status'] = 'Regular customer'
```

```python
# median days_since_prior_order less than 10 - label "Frequent customer"
ords_prods_merge.loc[ords_prods_merge['user_frequency'] <= 20, 'spend_status'] = 'Frequent customer'
```
