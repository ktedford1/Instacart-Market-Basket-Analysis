## Task questions

### Q2: Create a bar chart, a histogram, a scatterplot, and a line chart, using a sample subset for the line chart.

#### Bar Chart:

```python
# create a bar chart
bar1 = ords_prods['orders_time'].value_counts().sort_index().plot.bar(color = 'seagreen')
```

Note: THe number of orders peaks betwen 9 and 18 hours.

```python
# save the bar chart
bar1.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Q2_bar1.png'))
```

#### Histogram

```python
# create a histogram
hist1 = ords_prods['days_since_prior_order'].plot.hist(bins = 25, color = "teal")
```

Note: Most returning customers re-order on day 6, with another smaller peak on day 12. The remainders are binned in the final group.

```python
# save the histogram
hist1.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Q2_hist1.png'))
```

#### Scatterplot

```python
# create a scatterplot
scatter1 = sns.scatterplot(x = 'Age', y = 'Age',data = ords_prods)
```

Note: there are no outliers in the age variable, and all ages are represented.

```python
# save the scatter plot
scatter1.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Q2_scatter1.png'))
```

#### Line Chart

```python
# create a df with only the columns you need for a line chart (use the "df_test" df as a sample subset)
df_2 = df_test[['orders_day_of_week','prices']]
```

```python
# create a line chart from the df_2 test set
line1 = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

Note: There is a clear difference between this plot and the plot shown in the example that is likely due to the non-representative subset used here, df_test, which consists of only the first million rows of the dataframe.

```python
# save the line chart
line1.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'Q2_line1.png'))
```
