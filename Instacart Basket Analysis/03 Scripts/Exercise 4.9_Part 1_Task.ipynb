#### Data consistency checks: missing data, duplicated data, mixed type data

```python
# duplicated data check: no duplicates!
df_dups = customers[customers.duplicated()]
```

```python
df_dups
```

```python
# missing data check: no null values!
customers.isnull().sum()
```

```python
# mixed type data check
for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
  else:
    print("Consistent")
```

```python
# fix the datatype of the "First Name" column
customers['First Name'] = customers['First Name'].astype('str')
```
